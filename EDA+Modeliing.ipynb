# **Explanatory data analysis**

```python
def visualize_distribution(dataframe, column):
    plt.figure(figsize=(4, 3))

    # Distribution visualization
    if df[column].dtype == 'category':
        counts = df[column].value_counts().sort_index()
        labels = counts.index.astype(str)
        sizes = counts.values
        colors = sns.color_palette('Oranges', len(labels))
        patches, texts, autotexts = plt.pie(sizes, colors=colors, startangle=90,
                                            wedgeprops={'linewidth': 0.35, 'edgecolor': 'black'},
                                            autopct='%1.1f%%')

        # Customizing labels
        plt.gca().set_aspect('equal')
        plt.title(f'Distribution of {column}', fontsize=10)
        plt.legend(patches, labels, loc='upper right', bbox_to_anchor=(1.15, 1), frameon=True,
                   labelspacing=0.5, fontsize=6, title=None,
                   handletextpad=1, handlelength=0.8)

        # Add percentages to legend
        total = sum(sizes)
        percentages = [f'{(100 * size / total):.1f}%' for size in sizes]
        for i, text in enumerate(plt.gca().get_legend().get_texts()):
            text.set_text(f'{labels[i]} ({percentages[i]})')

        # Add percentages inside the pie chart (only for >= 10%)
        for autotext, size in zip(autotexts, sizes):
            if size / sum(sizes) >= 0.1:
                autotext.set_color('black')
                autotext.set_fontsize(8)
            else:
                autotext.set_text('')

    else:
        plt.grid(True, linestyle='dotted', linewidth=0.5, alpha=1, zorder=0)

        # Separate data based on target values
        data_target_0 = dataframe[dataframe['target'] == 0]
        data_target_1 = dataframe[dataframe['target'] == 1]

        # Plot histograms with different colors
        sns.histplot(data_target_0[column], bins=25, color='yellow', edgecolor='black', label='Less Change for HA')
        sns.histplot(data_target_1[column], bins=25, color='orange', edgecolor='black', label='More Change for HA')

        # Add dashed lines for outlier boundaries
        q1 = df[column].quantile(0.25)
        q3 = df[column].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr

        if lower_bound >= 0:
            plt.axvline(lower_bound, color='purple', linestyle='dashed', linewidth=0.75, label='Outlier Lower Bound')
        plt.axvline(upper_bound, color='blue', linestyle='dashed', linewidth=0.75, label='Outlier Upper Bound')

        plt.xlabel(column.capitalize(), fontsize=8)
        plt.ylabel('Count', fontsize=8)
        plt.title(f'Distribution of {column}', fontsize=10)
        plt.legend(fontsize=6)

    plt.tight_layout()
    plt.show()
```

# **Training for Logistic Regression using Manually Code**

Berikut adalah penjelasan matematika dari masing-masing rumus yang digunakan dalam model logistic regression:

1. **Fungsi Sigmoid**:
   $\sigma(z) = \frac{1}{1 + e^{-z}} $
   $\sigma(z) = \frac{\frac(e^{z}{e^{z}})}{1 + e^{-z}} $
   > Fungsi sigmoid digunakan untuk menghitung probabilitas bahwa suatu sampel masuk ke dalam kelas positif. Ini memetakan nilai dari domain yang luas menjadi nilai di antara 0 dan 1.

2. **Inisialisasi Parameter**:
    > $w$ adalah vektor bobot dengan dimensi $(n, 1)$, diinisialisasi secara acak dengan nilai kecil. $(b)$ adalah bias yang diinisialisasi dengan nilai nol.

3. **Forward Propagation**:
   $Z = w^T X + b,$
   $A = \sigma(Z)$
   > $(Z)$ adalah hasil dari perkalian matriks antara bobot $(w)$ (transpose) dan fitur $(X)$, ditambah dengan bias $(b )$. $(A)$ adalah hasil dari fungsi sigmoid yang diterapkan ke $(Z)$, yang menghasilkan nilai prediksi.

4. **Loss Function (Cost Function)**:
   $J(w, b) = -\frac{1}{m} \sum_{i=1}^{m} [y^{(i)}\log(\hat{y}^{(i)}) + (1-y^{(i)})\log(1-\hat{y}^{(i)})]$
   > Di mana $(m)$ adalah jumlah sampel, $(y^{(i)})$ adalah label aktual, dan $(\hat{y}^{(i)})$ adalah prediksi yang dihasilkan oleh model. Cost function ini adalah cross-entropy loss function yang umum digunakan dalam logistic regression.

5. **Backward Propagation**:
   $\frac{\partial J}{\partial w} = \frac{1}{m} X^T (\hat{y}^{(i)} - y^{(i)}),$
   $\frac{\partial J}{\partial b} = \frac{1}{m} \sum_{i=1}^{m} (\hat{y}^{(i)} - y^{(i)})$
   > Gradien dari loss function terhadap bobot dan bias. Ini adalah langkah untuk mengupdate parameter agar cost function menurun.

6. **Parameter Update**:
   $w = w - \alpha \frac{\partial J}{\partial w},$
   $b = b - \alpha \frac{\partial J}{\partial b}$
   > Di mana $(\alpha)$ adalah laju pembelajaran $(\alpha/ learning rate)$. Parameter diupdate menggunakan gradien yang telah dihitung sebelumnya.

7. **Prediksi**:
   $\hat{y}_{new} = \sigma(w \cdot x_{new} + b)$
   > Untuk membuat prediksi pada data baru, kita menggunakan fungsi sigmoid untuk menghitung probabilitas bahwa data tersebut masuk ke dalam kelas positif.


```python
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def logistic_regression(X, y, learning_rate=0.01, num_iterations=1000):
    num_samples, num_features = X.shape
    weights = np.zeros(num_features)
    bias = 0
    best_cost = float('inf')
    best_weights = None
    best_bias = None

    for i in range(num_iterations):
        # Forward propagation
        linear_model = np.dot(X, weights) + bias
        y_predicted = sigmoid(linear_model)

        # Compute cost
        cost = -1/num_samples * np.sum(y * np.log(y_predicted) + (1 - y) * np.log(1 - y_predicted))

        # Backward propagation
        dw = 1/num_samples * np.dot(X.T, (y_predicted - y))
        db = 1/num_samples * np.sum(y_predicted - y)

        # Update parameters
        weights -= learning_rate * dw
        bias -= learning_rate * db

        # Print cost and weights every 1000 iterations
        if i % 1000 == 0:
            print("Cost after iteration {}: {:.6f}".format(i, cost))
            print("Weights after iteration {}: {}".format(i, weights))

        # Save best cost and parameters
        if cost < best_cost:
            best_cost = cost
            best_weights = weights
            best_bias = bias

    return best_weights, best_bias, best_cost

# Contoh penggunaan:
learning_rate = 0.01
num_iterations = 100000
best_weights, best_bias, best_cost = logistic_regression(X_train, y_train, learning_rate, num_iterations)

print("\nBest Cost: {:.6f}".format(best_cost))
print("Best Weights: {}".format(best_weights))
print("Best Bias: {:.6f}".format(best_bias))

# Prediksi dengan menggunakan best_weights dan best_bias
linear_model_train = np.dot(X_train, best_weights) + best_bias
predictions_train = sigmoid(linear_model_train)
predictions_train_cls = [1 if i > 0.5 else 0 for i in predictions_train]

linear_model_test = np.dot(X_test, best_weights) + best_bias
predictions_test = sigmoid(linear_model_test)
predictions_test_cls = [1 if i > 0.5 else 0 for i in predictions_test]

# Menghitung akurasi
train_accuracy = np.sum(predictions_train_cls == y_train) / len(y_train)
test_accuracy = np.sum(predictions_test_cls == y_test) / len(y_test)

print("\nTrain accuracy: {:.2f}%".format(train_accuracy * 100))
print("Test accuracy: {:.2f}%".format(test_accuracy * 100))
```

# **Training for SVM (Linear) Using Manully Code**

```python
# Mengubah nilai y_train menjadi -1 dan 1
y_train = np.where(y_train != 1, -1, 1)
y_train
```

```python
# Mengubah nilai y_train menjadi -1 dan 1
y_test = np.where(y_test != 1, -1, 1)
y_test
```

```python
import numpy as np
from sklearn.metrics import accuracy_score

class LinearSVC:
    def __init__(self, learning_rate=0.01, num_iterations=1000):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.weights = None
        self.bias = None
        self.best_weights = None
        self.best_bias = None
        self.best_loss = float('inf')

    def fit(self, X, y):
        num_samples, num_features = X.shape
        self.weights = np.zeros(num_features)
        self.bias = 0

        # Mengubah label 0 menjadi -1
        y_modified = np.where(y <= 0, -1, 1)

        # Gradient Descent
        for iteration in range(self.num_iterations):
            # Compute gradients
            distances = 1 - y_modified * (np.dot(X, self.weights) + self.bias)
            dw = np.zeros(num_features)
            db = 0
            for idx, dist in enumerate(distances):
                if max(0, dist) == 0:
                    dw += self.weights
                else:
                    dw += self.weights - (y_modified[idx] * X[idx])
                    db += -y_modified[idx]

            dw /= num_samples
            db /= num_samples

            # Update parameters
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

            # Compute loss
            loss = np.sum(np.maximum(0, distances))

            # Print loss every 100 iterations
            if iteration % 100 == 0:
                print(f"Iteration {iteration}: \nLoss = {loss} \nWeight {self.weights} \nBias {self.bias}")

            # Save best parameters
            if loss < self.best_loss:
                self.best_loss = loss
                self.best_weights = self.weights.copy()
                self.best_bias = self.bias

    def predict(self, X):
        linear_model = np.dot(X, self.best_weights) + self.best_bias
        return np.where(linear_model <= 0, -1, 1)  # Mengubah output menjadi -1 dan 1

# Contoh penggunaan:
svm = LinearSVC()
svm.fit(X_train, y_train)

# Prediksi untuk data latih
train_predictions = svm.predict(X_train)

# Prediksi untuk data uji
test_predictions = svm.predict(X_test)

# Menghitung akurasi
train_accuracy = accuracy_score(y_train, train_predictions)
test_accuracy = accuracy_score(y_test, test_predictions)

# Print parameter terbaik
print("\nBest Weights:", svm.best_weights)
print("Best Bias:", svm.best_bias)

print("Train Accuracy:", train_accuracy)
print("Test Accuracy:", test_accuracy)
```

# **Training for Support Vector Machine (RBF) using Manually Code**

Dalam implementasi SVM , berikut adalah beberapa rumus matematika yang digunakan:

1. **Fungsi Keputusan (Decision Function):**
   $f(x) = \sum_{i=1}^{N} \alpha_i y_i K(x_i, x) + b$

2. **Fungsi Kernel RBF:**
   $K(x_i, x_j) = e^{-\gamma \lVert x_i - x_j \rVert^2}$

3. **Update Alpha (Alpha Update):**
   $\alpha_j^{\text{new}} = \alpha_j^{\text{old}} + \frac{y_j (E_i - E_j)}{\eta}$

4. **Batas untuk Alpha_j (Bounds for Alpha_j):**
   $L = \max(0, \alpha_j^{\text{old}} - \alpha_i^{\text{old}})$

   $H = \min(C, C + \alpha_j^{\text{old}} - \alpha_i^{\text{old}})$

5. **Update Bias (Bias Update):**
   $b_1 = b - E_i - y_i (\alpha_i^{\text{new}} - \alpha_i^{\text{old}}) K(x_i, x_i) - y_j (\alpha_j^{\text{new}} - \alpha_j^{\text{old}}) K(x_i, x_j)$

   $b_2 = b - E_j - y_i (\alpha_i^{\text{new}} - \alpha_i^{\text{old}}) K(x_i, x_j) - y_j (\alpha_j^{\text{new}} - \alpha_j^{\text{old}}) K(x_j, x_j)$

6. **Perhitungan Bias (Bias Computation):**
   $\text{bias} = \frac{b_1 + b_2}{2}$

7. **Prediksi (Prediction):**
   $\hat{y} = \text{sign}(\sum_{i=1}^{N_{\text{SV}}} \alpha_i y_i K(x_i, x) + \text{bias})$

Di mana:
- $(x )$ dan $( y )$ adalah data pelatihan.
- $( N )$ adalah jumlah sampel pelatihan.
- $( N_{\text{SV}} )$ adalah jumlah vektor pendukung.
- $( \alpha_i )$ adalah koefisien Lagrange.
- $( E_i )$ adalah error prediksi pada sampel $( i )$.
- $( K(x_i, x_j) )$ adalah nilai kernel antara sampel $( i )$ dan $( j )$.
- $( C )$ adalah parameter penalti.
- $( b )$ adalah bias.

```python
import numpy as np

class SVM:
    def __init__(self, C=1.0, kernel='rbf', gamma=1.0):
        self.C = C  # Parameter penalti
        self.kernel = kernel  # Jenis kernel
        self.gamma = gamma  # Parameter gamma untuk kernel RBF
        self.support_vectors = None
        self.support_vector_labels = None
        self.support_vector_alphas = None
        self.support_vector_bias = None

    def rbf_kernel(self, X1, X2):
        # Hitung kernel RBF antara dua set data X1 dan X2
        return np.exp(-self.gamma * np.linalg.norm(X1[:, np.newaxis] - X2, axis=2) ** 2)

    def fit(self, X, y):
        n_samples, n_features = X.shape

        # Hitung kernel
        if self.kernel == 'rbf':
            self.K = self.rbf_kernel(X, X)
        else:
            raise ValueError("Unsupported kernel type")

        # Inisialisasi alpha, bias, dan error cache
        self.alphas = np.zeros(n_samples)
        self.bias = 0.0
        self.error_cache = np.zeros(n_samples)

        # Training menggunakan algoritma SMO
        passes = 0
        while passes < 1000:
            num_changed_alphas = 0
            for i in range(n_samples):
                error_i = np.dot(self.alphas * y, self.K[:, i]) + self.bias - y[i]
                if (y[i] * error_i < -1E-7 and self.alphas[i] < self.C) or (y[i] * error_i > 1E-7 and self.alphas[i] > 0):
                    j = np.random.choice(np.delete(np.arange(n_samples), i))  # Pilih acak j ≠ i
                    error_j = np.dot(self.alphas * y, self.K[:, j]) + self.bias - y[j]

                    # Simpan alpha lama
                    alpha_i_old = self.alphas[i]
                    alpha_j_old = self.alphas[j]

                    # Hitung batas untuk alpha_j
                    if y[i] != y[j]:
                        L = max(0, self.alphas[j] - self.alphas[i])
                        H = min(self.C, self.C + self.alphas[j] - self.alphas[i])
                    else:
                        L = max(0, self.alphas[j] + self.alphas[i] - self.C)
                        H = min(self.C, self.alphas[j] + self.alphas[i])

                    if L == H:
                        continue

                    # Hitung konstanta eta
                    eta = 2 * self.K[i, j] - self.K[i, i] - self.K[j, j]
                    if eta >= 0:
                        continue

                    # Hitung alpha_j baru
                    self.alphas[j] -= y[j] * (error_i - error_j) / eta

                    # Potong alpha_j sesuai batas
                    self.alphas[j] = np.clip(self.alphas[j], L, H)

                    # Periksa apakah alpha_j berubah cukup signifikan
                    if np.abs(self.alphas[j] - alpha_j_old) < 1e-5:
                        continue

                    # Hitung alpha_i baru
                    self.alphas[i] += y[i] * y[j] * (alpha_j_old - self.alphas[j])

                    # Hitung bias
                    b1 = self.bias - error_i - y[i] * (self.alphas[i] - alpha_i_old) * self.K[i, i] - y[j] * (self.alphas[j] - alpha_j_old) * self.K[i, j]
                    b2 = self.bias - error_j - y[i] * (self.alphas[i] - alpha_i_old) * self.K[i, j] - y[j] * (self.alphas[j] - alpha_j_old) * self.K[j, j]

                    if 0 < self.alphas[i] < self.C:
                        self.bias = b1
                    elif 0 < self.alphas[j] < self.C:
                        self.bias = b2
                    else:
                        self.bias = (b1 + b2) / 2

                    num_changed_alphas += 1

            if num_changed_alphas == 0:
                passes += 1
            else:
                passes = 0

            if passes % 100 == 0:
                idx = np.where((self.alphas > 1e-15) & (self.alphas < self.C - 1e-15))[0]
                weight = np.sum(self.alphas[idx][:, np.newaxis] * y[idx][:, np.newaxis] * X[idx], axis=0)
                cost = np.sum(self.alphas) - 0.5 * np.dot(self.alphas, np.dot(self.K, self.alphas))
                print(f"Iteration {passes}: Weight = {weight}, Cost = {cost}, Bias = {self.bias}")

        # Simpan vektor pendukung
        support_vector_indices = np.where(self.alphas > 1e-7)[0]
        self.support_vectors = X[support_vector_indices]
        self.support_vector_labels = y[support_vector_indices]
        self.support_vector_alphas = self.alphas[support_vector_indices]
        print('Indikasi Support Vector: ',support_vector_indices)
        print('Kandidat Support Vector: ',self.support_vectors )
        print('Kandidat Support Vector Alpha: ',self.support_vector_alphas )

        # Hitung bias dari vektor pendukung
        self.support_vector_bias = np.mean(self.support_vector_labels - np.dot(self.support_vector_alphas * self.support_vector_labels, self.rbf_kernel(self.support_vectors, self.support_vectors)))

    def predict(self, X):
        if self.kernel == 'rbf':
            K_pred = self.rbf_kernel(X, self.support_vectors)
        else:
            raise ValueError("Unsupported kernel type")

        return np.sign(np.dot(self.support_vector_alphas * self.support_vector_labels, K_pred.T) + self.support_vector_bias)

# Contoh penggunaan
svm_model = SVM(C=1.0, kernel='rbf', gamma=0.1)
svm_model.fit(X_train, y_train)

train_predictions = svm_model.predict(X_train)
test_predictions = svm_model.predict(X_test)

train_accuracy = accuracy_score(y_train, train_predictions)
test_accuracy = accuracy_score(y_test, test_predictions)

print("Train Accuracy:", train_accuracy)
print("Test Accuracy:", test_accuracy)
```
